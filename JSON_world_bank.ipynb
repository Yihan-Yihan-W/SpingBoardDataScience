```python
#import libraries
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
#load word_bank_projects.jason
data=json.load((open('data/world_bank_projects.json')))
```

```python
#Question1
#normalize data 
country=json_normalize(data)["countryname"]
#Find the 10 countries with most projects
country.value_counts().head(10)
```

```python
#Question2
#Find the top 10 major project themes (using column 'mjtheme_namecode')
themes = json_normalize(data,"mjtheme_namecode")
counts=themes.code.value_counts().head(10)
counts
```

```python
#list the themecode and name
themename=themes.drop_duplicates(subset=None, keep='first', inplace=False).sort_values("name")
themename = themename.drop_duplicates(subset = "code", keep = "last", inplace = False)
```

```python
#rename coloumn name of counts
#add index(code) as another coloumn 
counts=pd.DataFrame(counts)
counts.columns = ['counts']
counts["code"]=counts.index
```

```python
#show both themecode and themenames by merge two dataframes
result=pd.merge(counts, themename, how='inner', on="code")
result.reindex(columns=["name","code","counts"])
```

```python
#Question3
#sort by code and name
#replace '' with np.nan, then backfill NaN 
themes = json_normalize(data,"mjtheme_namecode")
themes=themes.sort_values(["code","name"]).replace('', np.nan).bfill()
themes.sort_index()
```

```python
#count again top 10 project themes
#same anwser as counting code as previous 
counts2=themes.name.value_counts()
counts2
```

```python
#change to dataframe
counts2=pd.DataFrame(counts2)
counts2.columns = ["counts"]
counts2["name"]=counts2.index
```

```python
#merge two dataframes to show both themename and themecode
result2 = pd.merge(counts2, themename, how = "inner", on = "name")
result2.reindex(columns=["name","code","counts"]).head(10)
```
